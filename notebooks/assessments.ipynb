```python
import requests
import pandas as pd
from datetime import datetime
import pandas as pd
```

```python
n_assessments = requests.get(
    """
    https://phl.carto.com/api/v2/sql?q=SELECT count(*) FROM assessments
    """
).json()["rows"][0]["count"]

print(f"There are {n_assessments} assessments as of {datetime.now().isoformat()}")

n_assessments_not_null = requests.get(
    """
    https://phl.carto.com/api/v2/sql?q=SELECT count(*) FROM assessments where parcel_number is not null
    """
).json()["rows"][0]["count"]

print(f"There are {n_assessments_not_null} assessments as of {datetime.now().isoformat()}")

years = requests.get(
    """
    https://phl.carto.com/api/v2/sql?q=SELECT min(year) as start_year, max(year) as end_year FROM assessments
    """
).json()["rows"][0]

print(f"{years['start_year']} to {years['end_year']}")
```

```python
assessments = requests.get(
    """
    https://phl.carto.com/api/v2/sql?q=SELECT assess.market_value, assess.year, assess.*,
    ST_Y((opa.the_geom)) AS lat, ST_X((opa.the_geom)) AS lng
    FROM assessments assess, opa_properties_public opa 
    where assess.parcel_number = opa.parcel_number 
    order by opa.parcel_number, assess.year
    limit 10
    """
).json()['rows']
```
