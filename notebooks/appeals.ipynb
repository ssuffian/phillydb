```python
import requests
import pandas as pd
from datetime import datetime
import pandas as pd
```

```python
appeals_count = requests.get(
    """
https://phl.carto.com/api/v2/sql?q=SELECT appealtype, count(*)
FROM appeals group by appealtype order by count(*) desc
"""
).json()["rows"]
pd.DataFrame(appeals_count)
```

```python
appeals = requests.get(
    """
https://phl.carto.com/api/v2/sql?q=SELECT opa_owner, count(*) FROM appeals where appealtype = 'ZBA Permit Denial - Variance'
group by opa_owner order by count(*) desc
"""
).json()["rows"]
pd.DataFrame(appeals)
```
