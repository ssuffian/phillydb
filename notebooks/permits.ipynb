```python
import requests
import pandas as pd
from datetime import datetime
import pandas as pd
```

```python
earliest_permit = requests.get(
    """
https://phl.carto.com/api/v2/sql?q=SELECT permitissuedate FROM permits order by permitissuedate limit 1
"""
).json()["rows"][0]["permitissuedate"]

latest_permit = requests.get(
    """
https://phl.carto.com/api/v2/sql?q=SELECT permitissuedate FROM permits order by permitissuedate desc limit 1
"""
).json()["rows"][0]["permitissuedate"]

n_permits = requests.get(
    """
https://phl.carto.com/api/v2/sql?q=SELECT count(*) FROM permits
"""
).json()["rows"][0]["count"]
print(f"There are {n_permits} licenses between {earliest_permit} and {latest_permit}")
```

```python
permits = requests.get(
    """
https://phl.carto.com/api/v2/sql?q=SELECT permitdescription, count(*) 
FROM permits group by permitdescription order by count(*) desc
"""
).json()["rows"]
pd.DataFrame(permits)
```
